```python
        #Network Analysis In Python

import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()

        #Adding nodes in a Graph

G.add_nodes_from([1,2,3,4,5])

        #Adding edges to the Graph

G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(2,5)
G.add_edge(3,5)
G.node[1]['label']='Blue'
G.node[2]['label']='Red'
G.node[3]['label']='Yellow'
G.node[4]['label']='Pink'
G.node[5]['label']='Green'
nx.draw(G)
plt.show();
G.nodes(data=True)
```

```python
    #To FInd the Number of edges
    
G.edges()
```

```python
    #To Find the neighbors of a Particular node

list(G.neighbors(2))
```

```python
nx.degree_centrality(G)
```

```python
    #Defining a function to find all those nodes who has m neighbors


def nodes_with_m_neighbors(G,m):
    nodes=set()
    for n in G.nodes():
        if len(list(G.neighbors(n)))==m:
            nodes.add(n)
    return nodes

nodes_with_m_neighbors(G,2)
```

```python
degree=[len(list(G.neighbors(n))) for n in G.nodes()]
degree
```

```python
    #Plotting a histogram of the Degress centrality

import matplotlib.pyplot as plt
deg_cent=nx.degree_centrality(G)
plt.figure()
plt.hist(list(deg_cent.values()))
plt.show()
```

```python
    #Defining a function to check whether a path exists between two nodes or not

def path_exists(G,node1,node2):
    visited_nodes=set()
    queue=[node1]
    for node in queue:
        neighbors=G.neighbors(node)
        if node2 in neighbors:
            print("Yes!Path exists")
            return True
            break
        else:
            visited_nodes.add(node)
            queue.extend([G.neighbors(node) for n in neighbors if n not in visited_nodes])
            if node==queue[-1]:
                print("Path do not exists")
                return False
```

```python
path_exists(G,1,5)
```

```python
path_exists(G,1,3)
```
